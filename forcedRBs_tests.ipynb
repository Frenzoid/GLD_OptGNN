```python
import numpy as np
import matplotlib.pyplot as plt

# --- Your Reproduced OptGNN RB200 Results ---
mean_yours_rb200 = 1.0102
std_yours_rb200 = 0.0044

# --- Visually Estimated Paper OptGNN RB200 Results (from Figure 2b) ---
# !! Please refine these estimates by looking closely at the paper's plot !!
paper_optgnn_mean_rb200_est = 1.008  # Example visual estimate for the mean
paper_optgnn_std_rb200_est = 0.007   # Example visual estimate for std dev (half error bar)

# --- Data for Plotting ---
methods_to_plot = [
    'OptGNN (Paper Est. RB200)',
    'OptGNN (Yours RB200)'
]

plot_means = [
    paper_optgnn_mean_rb200_est,
    mean_yours_rb200
]
plot_stds = [
    paper_optgnn_std_rb200_est,
    std_yours_rb200
]

# --- Create the Plot ---
plt.style.use('seaborn-v0_8-whitegrid') # Using a seaborn style
fig, ax = plt.subplots(figsize=(8, 4)) # Adjusted figure size for two bars

bar_colors = ['lightcoral', 'skyblue'] # Color for Paper's estimate, Your result
y_pos = np.arange(len(methods_to_plot))

# Create horizontal bars
bars = ax.barh(y_pos, plot_means, xerr=plot_stds, align='center', ecolor='black', capsize=5, height=0.6)
bars[0].set_color(bar_colors[0])
bars[1].set_color(bar_colors[1])

ax.set_yticks(y_pos)
ax.set_yticklabels(methods_to_plot, fontsize=10)
ax.invert_yaxis()  # To have "Paper Est." on top
ax.set_xlabel('Approximation Ratio (values closer to 1 are better)', fontsize=11)
ax.set_title('OptGNN Min-Vertex-Cover Performance on RB200', fontsize=12, pad=15)

# Set x-axis limits to focus on the ratios around 1.0
# Dynamically calculate good limits based on data
all_values_for_limits = []
for mean, std in zip(plot_means, plot_stds):
    all_values_for_limits.append(mean - std)
    all_values_for_limits.append(mean + std)
all_values_for_limits.append(1.0) # Ensure 1.0 is included

x_min_limit = min(all_values_for_limits) - 0.002
x_max_limit = max(all_values_for_limits) + 0.005 # A bit more space for text labels
ax.set_xlim(x_min_limit, x_max_limit)

# Add a vertical line at ratio = 1.000 for reference
ax.axvline(1.0, color='grey', linestyle='--', linewidth=0.8, label='Optimal (Ratio 1.0)', zorder=0)

# Add text labels for mean values next to the bars
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width + 0.0005, # Position text slightly to the right of the bar
            bar.get_y() + bar.get_height()/2, # Vertically centered
            f'{width:.4f}', # Format the number
            ha='left', va='center', fontsize=9)

plt.legend(loc='lower right', fontsize=9)
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust rect to make space for title
plt.grid(axis='x', linestyle=':', alpha=0.7)
plt.grid(axis='y', linestyle='None') # Remove horizontal grid lines for cleaner look

plt.show()

# --- Textual Comparison ---
print("\n--- Numerical Comparison (RB200 Vertex Cover) ---")
print(f"Paper's OptGNN Est. Mean  Std: {paper_optgnn_mean_rb200_est:.4f}  {paper_optgnn_std_rb200_est:.4f}")
print(f"Your OptGNN Mean  Std:         {mean_yours_rb200:.4f}  {std_yours_rb200:.4f}")

mean_difference = mean_yours_rb200 - paper_optgnn_mean_rb200_est
print(f"Difference in Mean: {mean_difference:+.4f}")
if paper_optgnn_mean_rb200_est != 0:
    relative_diff = (mean_difference / paper_optgnn_mean_rb200_est) * 100
    print(f"Relative Difference in Mean: {relative_diff:+.2f}%")

print("\nConclusion:")
if abs(mean_difference) <= 0.003 and abs(std_yours_rb200 - paper_optgnn_std_rb200_est) <= 0.003:
    print("Your OptGNN results for RB200 are very closely comparable to the paper's estimated performance.")
elif mean_yours_rb200 >= paper_optgnn_mean_rb200_est - paper_optgnn_std_rb200_est and \
     mean_yours_rb200 <= paper_optgnn_mean_rb200_est + paper_optgnn_std_rb200_est:
    print("Your OptGNN mean ratio falls within the estimated error margin (mean  std) of the paper's OptGNN.")
else:
    print("There are some differences between your OptGNN results and the paper's estimates. This could be due to visual estimation inaccuracies, hyperparameter differences, or minor implementation details.")
```
